## ðŸ“Š Model Quality Visualization

The chart below summarizes average LLM performance across evaluation metrics.  
Models are sorted by overall score, and differences in clarity or engagement can be inspected in the multi-metric breakdown.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load labeled data
df = pd.read_csv("/Users/michaelromero/Documents/AI-Evaluation-Demo/evaluations/labeled_responses.csv", encoding='cp1252')

# Compute averages by model
avg_scores = (
    df.groupby("model")["overall_score"]
    .mean()
    .sort_values(ascending=False)
)

# Create figure
plt.figure(figsize=(8,5))
bars = plt.bar(avg_scores.index, avg_scores.values, color="#4F86F7", edgecolor="black", linewidth=0.8)

# Add numeric labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.05,
        f"{height:.2f}",
        ha="center",
        va="bottom",
        fontsize=10,
        weight="bold"
    )

# Titles & labels
plt.title("Average LLM Quality Scores by Model", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Model", fontsize=11, labelpad=10)
plt.ylabel("Average Score (0â€“5)", fontsize=11, labelpad=10)

# Formatting tweaks
plt.ylim(0, 5.5)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.xticks(rotation=15)
plt.tight_layout()

plt.show()
```
