```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the labeled responses
df = pd.read_csv("/Users/michaelromero/Documents/AI-Evaluation-Demo/evaluations/labeled_responses.csv", encoding='cp1252')

# Preview data
print("Sample of data:")
display(df.head())

# Average score by model
avg_scores = df.groupby("model")["overall_score"].mean()
print("\nAverage Overall Score by Model:\n", avg_scores)

# Visualize
plt.figure(figsize=(6,4))
avg_scores.plot(kind="bar")
plt.title("Average LLM Quality Scores")
plt.ylabel("Overall Score (out of 5)")
plt.xlabel("Model")
plt.ylim(0,5)
plt.show()

# Identify lowest-scoring responses
lowest = df.sort_values(by="overall_score").head(3)
print("\nLowest Scoring Responses:\n")
display(lowest[["id", "model", "response", "overall_score"]])
```
